```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
P0 = 100
K = 1000
r = 0.1
t = np.linspace(0, 100, 400)  # time from 0 to 100 with 400 points

# Logistic growth function
P = K / (1 + (K - P0) / P0 * np.exp(-r * t))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, P, label='Logistic Growth Curve')
plt.title('Cell Population Growth')
plt.xlabel('Time (t)')
plt.ylabel('Population (P)')
plt.legend()
plt.grid(True)
plt.show()
```
